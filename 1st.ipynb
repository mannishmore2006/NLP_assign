```python
# Import the Natural Language Toolkit library
import nltk

# Download tokenizer models
nltk.download('punkt')

# Download WordNet database for lemmatization
nltk.download('wordnet')

# Download additional WordNet data
nltk.download('omw-1.4')
```

```python
# Sample text for NLP operations
text = "NLTK is amazing! I love learning NLP, and I'm studying multi-word expressions like New York City ðŸ˜Š #AI"

# Display the text
text
```

```python
# Import WhitespaceTokenizer
from nltk.tokenize import WhitespaceTokenizer

# Create tokenizer object
wt = WhitespaceTokenizer()

# Split text based only on spaces
wt.tokenize(text)
```

```python
# Import punctuation tokenizer
from nltk.tokenize import wordpunct_tokenize

# Split words and punctuation separately
wordpunct_tokenize(text)
```

```python
# Import Treebank tokenizer
from nltk.tokenize import TreebankWordTokenizer

# Create Treebank tokenizer object
tb = TreebankWordTokenizer()

# Tokenize text using grammar-based rules
tb.tokenize(text)
```

```python
# Import TweetTokenizer
from nltk.tokenize import TweetTokenizer

# Create tweet tokenizer object
tt = TweetTokenizer()

# Tokenize text containing hashtags and emojis
tt.tokenize(text)
```

```python
# Import MWE tokenizer
from nltk.tokenize import MWETokenizer

# Define multi-word expressions
mwe = MWETokenizer([('New', 'York', 'City')], separator='_')

# Tokenize text while keeping multi-word terms together
mwe.tokenize(text.split())
```

```python
# Tokenize text using punctuation tokenizer
tokens = wordpunct_tokenize(text)

# Display tokens
tokens
```

```python
# Import Porter Stemmer
from nltk.stem import PorterStemmer

# Create stemmer object
ps = PorterStemmer()

# Apply stemming to each token
[ps.stem(word) for word in tokens]
```

```python
# Import Snowball Stemmer
from nltk.stem import SnowballStemmer

# Create English Snowball stemmer
ss = SnowballStemmer("english")

# Apply Snowball stemming
[ss.stem(word) for word in tokens]
```

```python
# Import WordNet Lemmatizer
from nltk.stem import WordNetLemmatizer

# Create lemmatizer object
lemmatizer = WordNetLemmatizer()

# Convert words into meaningful base form
[lemmatizer.lemmatize(word) for word in tokens]
```
